```python
import pandas as pd
import hvplot
import hvplot.pandas as hvp
```

```python
import calendar

# list of month names using calendar module
list_of_months = list(calendar.month_name)[1:]
```

```python
list_df = []

for month in list_of_months:
    month_df = pd.read_csv(f'{month}.csv')
    month_df['month'] = month
    list_df.append(month_df)

year_df = pd.concat(list_df)
```
