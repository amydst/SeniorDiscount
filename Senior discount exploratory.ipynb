```python
import pandas as pd
import hvplot
import hvplot.pandas as hvp
```

```python
import calendar
year = 2022

# list of month names using calendar module
list_of_months = list(calendar.month_name)[1:]

list_of_months
```

```python
list_df = []

january = pd.read_csv(f'2022/January.csv')

january
```

```python


for month in list_of_months:
    month_df = pd.read_csv(f'{year}\{month}.csv')
    month_df['month'] = month
    list_df.append(month_df)

year_df = pd.concat(list_df)
```

```python
year_df['StartTime'] = pd.to_datetime(year_df['StartTime'])

year_df['EndTime'] = pd.to_datetime(year_df['EndTime'])
```

```python
year_df = year_df.drop(['Department'],axis=1)

year_df.head(10)
```

```python

year_df_grouped = year_df.groupby('TRN_PK').agg(
    TotalDiscountAmount = ('DiscountAmount','sum'),
    Start = ('StartTime', 'last'),
    End = ('EndTime', 'last'),
    ReceiptNumber = ('ReceiptNumber', 'first'),
    CustomerAccount = ('CustomerAccount','first'),
    TransactionTotal = ('TransactionTotal', 'first'),
    NetSales = ('NetSales', 'first'),
    TotalRedemptionCount = ('RedemptionCount', 'sum'),
    TotalTLIqty = ('TLI_Quantity','sum'),
    TotalTLIamt = ('TLI_Amount','sum'),
    Month = ('month','first')
).reset_index()

```

```python
year_df_grouped.tail(5)
```

```python
year_df_grouped = year_df_grouped.assign(
    EndDate=year_df_grouped["End"].dt.date, 
    EndTime=year_df_grouped["End"].dt.time
)

```

```python
year_df_grouped['TotalDiscountAmount'] = year_df_grouped['TotalDiscountAmount']* -1
```

```python
year_df_grouped.tail(5)
```

```python
by_day = year_df_grouped.groupby('EndDate').agg(
    CustomerCount = ('TRN_PK','count'),
    TotalDiscountAmount = ('TotalDiscountAmount','sum'),
    CustomerAccount = ('CustomerAccount','count'),
    TransactionTotal = ('TransactionTotal', 'sum'),
    TotalRedemptionCount = ('TotalRedemptionCount', 'sum'),
    Month = ('Month','first')
).reset_index()

by_day['year'] = year
by_day.head(5)   
```
